## Sample Records

```python
df_sample = df.sample(n=500000, random_state=42)

print("Sample created ✅")
print(df_sample.shape)
```

```python
df_sample.head()
```

## Remove Duplicates

```python
before = df_sample.shape[0]

df_sample = df_sample.drop_duplicates()

after = df_sample.shape[0]

print("Duplicates removed:", before - after)
```

## Handle missing values

```python
df_sample = df_sample.dropna(subset=["Date", "Primary Type", "Latitude", "Longitude"])

print("After dropping critical nulls:")
df_sample.isnull().sum()
```

#### Drop rows where location is missing

```python
df_sample = df_sample.dropna(subset=["Latitude", "Longitude"])

print("After dropping location nulls:")
df_sample.shape
```

#### Filling less important missing values

```python
# Location Description → fill with 'UNKNOWN'
df_sample["Location Description"] = df_sample["Location Description"].fillna("UNKNOWN")
```

```python
# Ward & Community Area → fill with mode (most frequent value)
df_sample["Ward"] = df_sample["Ward"].fillna(df_sample["Ward"].mode()[0])
df_sample["Community Area"] = df_sample["Community Area"].fillna(df_sample["Community Area"].mode()[0])
```

```python
# District (few missing)
df_sample["District"] = df_sample["District"].fillna(df_sample["District"].mode()[0])
```

```python
df_sample.isnull().sum()
```

## Fix Data Types Properly

```python
# Convert Date
df_sample["Date"] = pd.to_datetime(df_sample["Date"], errors="coerce")
```

```python
# Convert categorical columns
categorical_cols = [
    "Primary Type",
    "Description",
    "Location Description",
    "IUCR",
    "FBI Code"
]

for col in categorical_cols:
    df_sample[col] = df_sample[col].astype("category")
```

```python
# Convert numeric columns properly
df_sample["District"] = df_sample["District"].astype(int)
df_sample["Ward"] = df_sample["Ward"].astype(int)
df_sample["Community Area"] = df_sample["Community Area"].astype(int)
```

## Drop Unnecessary Columns

```python
columns_to_drop = [
    "Case Number",
    "Block",
    "IUCR",
    "FBI Code",
    "Updated On",
    "Location"
]

df_sample = df_sample.drop(columns=columns_to_drop)

df_sample.columns
```
